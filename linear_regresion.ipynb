```python
## Linerar Regression
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
```

```python
import warnings 
warnings.filterwarnings("ignore")
```

```python
 model1.summary()
```

**Observations**
- The predicted Equation is y_hat = Beta_0 + Beta_1
- y_hat = 13.8356 + 1.2297
- The probability(p-value) for intercept (beta_0) is 0.707 > 0.05
- Therefore the intercept cofficent may not be that much significent in prediction
- However the p-value for "daily" (beta_1) is 0.00 < 0.05
- Therefore the beta_1 coefficient is highly significent and is contributint to prediction.

```python
## predict on givenall training data

pred = model1.predict(data["daily"])
pred
```

```python
data["Y_hat"] = pred
data
```

```python
data["residuals"]= data["sunday"]-data["Y_hat"]
data
```

```python
mse = np.mean((data["daily"]-data["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
### Plot the residuals versus y_hat (to check wheather residuals are independent of each other)
plt.scatter(data['Y_hat'],data["residuals"])
```

#### Observaions:
- There appers to be no trend and the resido=uals are randomly placed around the zero error line
- Hence the assupmtion of homoscedasticty is satisfied(constant variance in residuals)

```python
import statsmodels.api as sm
sm.qqplot(data["residuals"], line='45', fit=True)
plt.show
```
